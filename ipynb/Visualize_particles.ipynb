```python
from ecell4.core import *
from ecell4.bd import BDWorld
from ecell4.lattice import LatticeWorld
```

```python
# w = LatticeWorld(Position3(1, 1, 1), voxel_radius=0.0025)
w = BDWorld(Position3(1, 1, 1))
```

```python
w.add_molecules(Species("A", "0.0025", "1"), 60)
w.add_molecules(Species("B", "0.005", "0.5"), 60)
```

```python
print w.num_molecules(Species("A")), w.num_molecules(Species("B"))
```
